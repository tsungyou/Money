```python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq
import os
from datetime import datetime, timedelta
import yfinance as yf
CUR_PATH = os.getcwd()
DIR_PATH = os.path.join(os.path.dirname(CUR_PATH), "tw-stock-screener")
print(DIR_PATH)
today = datetime.now()

def get_files_under_path(path):
    files = []

    for root, directories, filenames in os.walk(path):
        for filename in filenames:
            file_path = os.path.join(root, filename)
            if os.path.isfile(file_path):
                files.append(file_path)

    return files

# Example usage:
path_parquet = os.path.join(DIR_PATH, "database")
print(path_parquet)
files = get_files_under_path(path_parquet)
for _, file in enumerate(files[:-1]):
    print("file", file)
    tick = file.split("/")[-1].split(".")[0]
    ticker = tick.replace("_", ".").split("\\")[-1]
    print("ticker", ticker)
    print("tick", tick)
    update_data =  yf.download(ticker, start=today-timedelta(days=3), interval='1d', progress=False)
    current_data = pd.read_parquet(file)
    df = pd.concat([current_data, update_data], axis=0)

    table = pa.Table.from_pandas(df)

    df.to_parquet(file)
    
    if _%100==0:
        print(_)
```

```python
from functions import data_strat2, data_strat1, nav_returns_backtest_only
import pandas as pd
stock_list = pd.read_csv("database/yearly/tw_stock_list.csv")
s = stock_list.sort_values(by="volume", ascending=False)
from datetime import datetime
df_final = pd.DataFrame()
start='2023-01-01'
end= datetime.now()
returns = []
winrates = []
avg_ret = []
trade_total = []
last_open_date = []
last_open_price = []
tickers = []
for ticker in s['market'][:500]:
    ticker_bar = ticker.replace(".", "_")
    df = data_strat1(ticker_bar, start, end)
    
    return_total, winrate, avg_return, trades, last_openposition_date, last_openposition_price = nav_returns_backtest_only(df)
    returns.append(return_total)
    winrates.append(winrate)
    avg_ret.append(avg_return)
    trade_total.append(trades)
    tickers.append(ticker[:4])
    last_open_date.append(df.index[last_openposition_date])
    last_open_price.append(last_openposition_price)
df_final['ticker'] = tickers
df_final['return_total'] = returns
df_final['winrate'] = winrates
df_final['avg_return'] = avg_ret
df_final['trade_total'] = trade_total
df_final['last_openposition_date'] = last_open_date
df_final['last_openposition_price'] = last_open_price


df_strat2 = pd.DataFrame()
start='2023-01-01'
end= datetime.now()
returns = []
winrates = []
avg_ret = []
trade_total = []
last_open_date = []
last_open_price = []
tickers = []
for ticker in s['market'][:500]:
    ticker_bar = ticker.replace(".", "_")
    df = data_strat2(ticker_bar, start, end)

    return_total, winrate, avg_return, trades, last_openposition_date, last_openposition_price = nav_returns_backtest_only(df)
    returns.append(return_total)
    winrates.append(winrate)
    avg_ret.append(avg_return)
    trade_total.append(trades)
    tickers.append(ticker[:4])
    last_open_date.append(df.index[last_openposition_date])
    last_open_price.append(last_openposition_price)
df_strat2['ticker'] = tickers
df_strat2['return_total'] = returns
df_strat2['winrate'] = winrates
df_strat2['avg_return'] = avg_ret
df_strat2['trade_total'] = trade_total
df_strat2['last_openposition_date'] = last_open_date
df_strat2['last_openposition_price'] = last_open_price
```
