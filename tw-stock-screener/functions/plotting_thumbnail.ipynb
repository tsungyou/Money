```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://histock.tw/twstock'
```

```python
import json 
res = requests.get("https://histock.tw/twclass/A034")
soup = BeautifulSoup(res.text)
with open("stock_classification.json", "r") as f:
    data_json = json.load(f)

ticker = soup.find_all("span", {'class':"w60 lft-p stockno"})
codename = soup.find_all("span", {"class":"w80 lft-p"})
new_dict = {ticker[i].text:codename[i].text for i in range(1, len(ticker))}
data_json['旅遊'] = new_dict
with open("stock_classification.json", "w") as f:
    json.dump(data_json, f, indent=4)
data_json['旅遊']
```

```python
stocks = list(data_json['旅遊'].keys())
stocks
```

```python
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
class_ = "旅遊"
stocks = list(data_json[class_].keys())
market = "TW"
stocks = [f"{stock}.{market}" for stock in stocks]

total = len(stocks) - 1
x = np.sqrt(total)
y = total/x
if y == np.round(y):
    y = int(y)
else:
    y = int(y) + 1
if x == np.round(x):
    x = int(x) + 1
else:
    x = int(x) + 1
print(x, y)
fig, ax = plt.subplots(x, y, figsize=(16,16))
a = yf.download(stocks, start='2023-01-01')['Close']
# print(a)
stock_iter = 0
for i in range(x):
    for j in range(y):
        ax[i, j].plot(a[stocks[stock_iter]])
        ax[i, j].set_title(stocks[stock_iter][:4])
        ax[i, j].set_xticklabels([])
        stock_iter += 1
plt.tight_layout()
plt.show()
```
