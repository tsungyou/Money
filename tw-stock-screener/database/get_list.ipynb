```python
import requests
import pandas as pd
import yfinance as yf
res = requests.get("https://histock.tw/stock/rank.aspx?p=all")
dfv = pd.read_html(res.text)[0]
dfv = dfv.iloc[:, [0, 1, 2, 11, 12]]
dfv.columns = ['ticker', "codename", "price", "volume", "volume_ntd"]
```

```python
import requests
import pandas as pd
import yfinance as yf
res = requests.get("https://histock.tw/stock/rank.aspx?p=all")
df = pd.read_html(res.text)[0]
df
# # 設定column名稱
# df.columns = df.iloc[0]
# # 刪除第一行
# df = df.iloc[1:]
df = df.iloc[:, [0, 1]]
df.columns = ['ticker', "codename"]
four_digit = df['ticker'].apply(lambda x: len(x) == 4)
df_4 = df[four_digit]

```

```python

with_market = []
for i in range(len(df_4)):
    ticker = df_4['ticker'].iloc[i]
    df = yf.download(f"{ticker}.TW", start='2024-04-29')
    if len(df) == 0:
        with_market.append(f"{ticker}.TWO")
    else:
        with_market.append(f"{ticker}.TW")
```

```python
# market 1234.TW/1234.TWO
df_4['market'] = with_market
# listed True/False
df_4['listed'] = df_4["market"].apply(lambda x: x[-1] == "W")
# TW_TWO TW/TWO
df_4['TW_TWO'] = df_4['listed'].apply(lambda x: "TW" if x else "TWO")
```

```python
df_new = pd.concat([dfv, df_4], axis=1, join='inner')
df_new.to_csv("tw_stock_list.csv")
```
