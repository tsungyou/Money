```python
import pandas as pd
from functions import data_strat1, nav_returns_backtest_only
stock_list = pd.read_csv("../database/yearly/tw_stock_list.csv")
start='2020-01-01'
end='2021-01-01'
df_final = pd.DataFrame()
stock_list = stock_list.sort_values(by='volume_ntd', ascending=False)
print(stock_list)
for ticker in stock_list['market'][:10]:
    ticker_bar = ticker.replace(".", "_")
    df = data_strat1(f"{ticker_bar}", start, end)

    return_total, winrate, avg_return, trades, last_openposition_date, last_openposition_price = nav_returns_backtest_only(df)
    df_final[ticker] = [return_total, winrate, avg_return, trades, df.index[last_openposition_date], last_openposition_price]

df_final.index = ["return_total", "winrate", "avg_return", "number_of_trades", "last_openposition_date", "last_openposition_price"]
```
