```python
import customtkinter as ctk

class App(ctk.CTk):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)

        self.title("Stock Manager")
        self.geometry("1920x1080")

        # Define menu options
        menu1 = ["Home", "Stock Info", "Portfolio", "Settings"]

        # Create an OptionMenu
        self.options = ctk.CTkOptionMenu(self, values=menu1, command=self.pages_selector)
        self.options.pack(pady=20)

        # Create a frame to hold the page content
        self.page_frame = ctk.CTkFrame(self)
        self.page_frame.pack(expand=True, fill='both')

        # Initialize the first page
        self.pages_selector(menu1[0])

    def pages_selector(self, selection):
        # Clear the current page frame
        for widget in self.page_frame.winfo_children():
            widget.destroy()

        # Load the selected page
        if selection == "Home":
            self.load_home_page()
        elif selection == "Stock Info":
            self.load_stock_info_page()
        elif selection == "Portfolio":
            self.load_portfolio_page()
        elif selection == "Settings":
            self.load_settings_page()

    def load_home_page(self):
        label = ctk.CTkLabel(self.page_frame, text="Welcome to the Home Page")
        label.pack(pady=20)

    def load_stock_info_page(self):
        label = ctk.CTkLabel(self.page_frame, text="Welcome to the Stock Info Page")
        label.pack(pady=20)

    def load_portfolio_page(self):
        label = ctk.CTkLabel(self.page_frame, text="Welcome to the Portfolio Page")
        label.pack(pady=20)

    def load_settings_page(self):
        label = ctk.CTkLabel(self.page_frame, text="Welcome to the Settings Page")
        label.pack(pady=20)

if __name__ == "__main__":
    app = App()
    app.mainloop()
```

```python
import customtkinter as ctk

class HomePage(ctk.CTkFrame):
    def __init__(self, parent):
        super().__init__(parent)
        label = ctk.CTkLabel(self, text="Welcome to the Home Page")
        label.pack(pady=20)

class StockInfoPage(ctk.CTkFrame):
    def __init__(self, parent):
        super().__init__(parent)
        label = ctk.CTkLabel(self, text="Welcome to the Stock Info Page")
        label.pack(pady=20)

class PortfolioPage(ctk.CTkFrame):
    def __init__(self, parent):
        super().__init__(parent)
        label = ctk.CTkLabel(self, text="Welcome to the Portfolio Page")
        label.pack(pady=20)

class SettingsPage(ctk.CTkFrame):
    def __init__(self, parent):
        super().__init__(parent)
        label = ctk.CTkLabel(self, text="Welcome to the Settings Page")
        label.pack(pady=20)

class App(ctk.CTk):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)

        self.title("Stock Manager")
        self.geometry("1920x1080")

        # Create a Tabview
        self.tabview = ctk.CTkTabview(self)
        self.tabview.pack(expand=True, fill='both')

        # Add tabs
        self.tabview.add("Home")
        self.tabview.add("Stock Info")
        self.tabview.add("Portfolio")
        self.tabview.add("Settings")

        # Create and place pages inside tabs
        self.home_page = HomePage(self.tabview.tab("Home"))
        self.home_page.pack(expand=True, fill='both')

        self.stock_info_page = StockInfoPage(self.tabview.tab("Stock Info"))
        self.stock_info_page.pack(expand=True, fill='both')

        self.portfolio_page = PortfolioPage(self.tabview.tab("Portfolio"))
        self.portfolio_page.pack(expand=True, fill='both')

        self.settings_page = SettingsPage(self.tabview.tab("Settings"))
        self.settings_page.pack(expand=True, fill='both')

if __name__ == "__main__":
    app = App()
    app.mainloop()
```
